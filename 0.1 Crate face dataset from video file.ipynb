# Create Dataset from webcam 
The following code takes the webcam output and collects faceimages.

```python
import cv2
import numpy as np
import os,datetime,time , sys
print ("Package loaded") 

ts = time.time()
# Convert time (in seconds) into a string
date = datetime.datetime.fromtimestamp(ts).strftime('%Y-%m-%d %H_%M_%S')
# create Folder with Timestamp
os.mkdir('images_ %s' %date)



# Capture Video using Webcam
cap=cv2.VideoCapture(0)

# Load CascadeClassifier to detect Faces -> frontalface.default.xml file)
face_cascade=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')




imageCount=0
saveCount=0
piclimit = 3000

while(saveCount<piclimit):
    #capture Frame by Frame
    ret,frame=cap.read()

    # convert frame to frame_gray -> Frame loses its colore and gets gray
    #frame_gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)

    #Detect face in the gray frame
    faces=face_cascade.detectMultiScale(frame)

    faceCaptured=0
    # defines a list called crop_img 
    crop_img=[None]


    for(x,y,w,h) in faces:	
        if((w>100)and(h>100)):
            cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
            faceCaptured+=1
            crop_img.append(cv2.resize(frame[y+2:y+h-1,x+2:x+w-1],(250,250)))



    cv2.imshow('frame',frame)

    if (faceCaptured>0):
        if (imageCount==0):
            cv2.imwrite('images_ %s/' %date +str(saveCount)+'.png',crop_img[1])
            saveCount +=1
            print("Number of saved pictures: [%s von %d]" %(saveCount,piclimit))
        if imageCount<2:
            imageCount +=1

        else :
            imageCount=0

    cv2.waitKey(1)

cap.release()
cv2.destroyAllWindows()

```

# Create Image Data from single images

The following code opens images and saves the face on the image in a different folder
